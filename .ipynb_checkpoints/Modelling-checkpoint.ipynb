## Load and Process Toronto Island Ferry Tickets Dataset

```python
def load_data(file):
    df = pd.read_csv(file,
                     dtype={'_id': int, 'Redemption Count': int, 'Sales Count': int},
                     parse_dates=['Timestamp'])

    df.sort_values('Timestamp', inplace=True)
    df.set_index('Timestamp', inplace=True)

    # Resample to daily frequency
    df_daily = df.resample('D').sum()
    
    # Rename to match model expectation
    df_daily.rename(columns={'Redemption Count': 'redemptions'}, inplace=True)

     # Add time-based features
    df_daily['day_of_week'] = df_daily.index.dayofweek      # 0 = Monday
    df_daily['weekofyear'] = df_daily.index.isocalendar().week
    df_daily['month'] = df_daily.index.month
    df_daily['quarter'] = df_daily.index.quarter
    df_daily['day_of_year'] = df_daily.index.dayofyear
    df_daily['is_weekend'] = df_daily.index.dayofweek.isin([5, 6]).astype(int)  # Saturday/Sunday

    # Lag features (redemptions and sales)
    df_daily['lag_1'] = df_daily['redemptions'].shift(1)
    df_daily['lag_7'] = df_daily['redemptions'].shift(7)
    df_daily['lag_14'] = df_daily['redemptions'].shift(14)

    df_daily['sales_lag_1'] = df_daily['Sales Count'].shift(1)
    df_daily['sales_lag_7'] = df_daily['Sales Count'].shift(7)

    # Rolling averages
    df_daily['rolling_mean_3'] = df_daily['redemptions'].rolling(window=3).mean()
    df_daily['rolling_mean_7'] = df_daily['redemptions'].rolling(window=7).mean()

    # Percent change (trend indicator)
    df_daily['pct_change'] = df_daily['redemptions'].pct_change()

    # Cumulative total
    df_daily['cumulative_redemptions'] = df_daily['redemptions'].cumsum()

    # Fill missing values (after lags/rolling)
    df_daily.fillna(0, inplace=True)

    return df_daily

# Load data
file_path = r"C:\Users\jiwue\Downloads\Toronto Island Ferry Ticket Counts.csv"
df = load_data(file_path)

df.head(10)
```

## Description of Features
- **_Id:** 	Unique identifier for each record

- **Redemptions:**	Number of ferry tickets redeemed (used for model prediction).

- **Sales Count:**	Number of ferry tickets sold on each day.

- **Day_of_week:**	Numeric representation of the day (0 = Monday, 6 = Sunday).

- **Weekofyear:**	Week number in the year (1â€“52).

- **Month:**  Month of the year (1â€“12).

- **Quarter:**	Quarterly calendar (1â€“4).

- **Day_of_year:**	Position of the day within the year (1â€“365).
                                                         
- **Is_weekend:**	Binary flag (1 = Saturday/Sunday, 0 = weekday).

- **Lag_1:**  Redemption count from 1 day ago.

- **Lag_7:**  Redemption count from 7 days ago.

- **Lag_14:** Redemption count from 14 days ago.
    
- **Sales_lag_1:**	Ticket sales from 1 day ago.

- **Sales_lag_7:**	Ticket sales from 7 days ago.

- **Rolling_mean_3:**  3-day moving average of redemptions.
    
- **Rolling_mean_7:**  7-day moving average.
    
- **PCT_change:**  Daily percentage change in redemptions.
    
- **Cumulative_redemptions:**  Total number of redemptions up to the present day

- **Timestamp**	 Original date of the observation

## Technical Description
I developed a forecasting system designed to predict ferry ticket redemptions and sales using a combination of traditional statistical methods and modern machine learning techniques. The dataset used is the Toronto Island Ferry Ticket Counts.csv, containing historical daily counts of ferry ticket redemptions and sales. The systemâ€™s goal is to capture seasonal patterns and trends in the data to improve the accuracy of future demand predictions.

### Modeling Approaches
The system compares four different models:

1.	**Base Model (Seasonal Decomposition):** This statistical approach decomposes the time series into seasonal components and leverages average seasonal patterns to generate forecasts. It serves as a simple baseline to understand underlying seasonality.

2. **Base Model Two (Trend + Seasonal Decomposition):** combines trend and seasonal components from seasonal decomposition. This model better accounts for long-term shifts in demand, leading to improved accuracy over the base model.

3. **Sales Model:** Specifically targets sales counts rather than redemptions, applying the same trend and seasonal decomposition approach to capture patterns relevant to sales volume.

4. **XGBoost Machine Learning Model:** A gradient boosting regression model that incorporates engineered features such as lag values (1, 7, 14 days), rolling averages, and calendar features (day of week, month, weekend indicator). This model leverages past trends, weekly and monthly seasonality, and holidays to make sharper, data-driven predictions.

### Evaluation Procedure
Using a time-series cross-validation strategy, the data was split into four sequential folds with a test size of one year (365 days) each. Each model was trained and tested on these splits to assess robustness and consistency. Performance was evaluated using three metrics:

- **Mean Absolute Percentage Error (MAPE):** Measures prediction accuracy as a percentage.
  
- **Root Mean Squared Error (RMSE):** Measures absolute error magnitude.

- **R-squared (RÂ²):** Indicates the proportion of variance explained by the model.

### Results Summary
- The Base Model showed limited predictive power with average MAPE of 86.5%, RMSE around 3769, and RÂ² of 0.17, highlighting the insufficiency of pure seasonal averages.

- The Base Model Two improved accuracy significantly (MAPE 48%, RMSE ~2860, RÂ² ~0.50), showing that incorporating trend components better captures demand changes.

- The Sales Model performed moderately well on sales forecasting, averaging a MAPE of 57.5%, RMSE ~2530, and RÂ² ~0.58, suitable for sales-specific planning.

- The XGBoost model outperformed all others by a large margin with average MAPE of just 29.7%, RMSE ~1342, and a high RÂ² of 0.89. This demonstrates the power of machine learning with engineered temporal and calendar features in capturing complex demand dynamics.
