## Load and process Toronto Island Ferry Tickets Dataset

```python
def load_data(file):
    df = pd.read_csv(file,
                     dtype={'_id': int, 'Redemption Count': int, 'Sales Count': int},
                     parse_dates=['Timestamp'])

    df.sort_values('Timestamp', inplace=True)
    df.set_index('Timestamp', inplace=True)

    # Resample to daily frequency
    df_daily = df.resample('D').sum()
    
    # Rename to match model expectation
    df_daily.rename(columns={'Redemption Count': 'redemptions'}, inplace=True)

     # Add time-based features
    df_daily['day_of_week'] = df_daily.index.dayofweek      # 0 = Monday
    df_daily['weekofyear'] = df_daily.index.isocalendar().week
    df_daily['month'] = df_daily.index.month
    df_daily['quarter'] = df_daily.index.quarter
    df_daily['day_of_year'] = df_daily.index.dayofyear
    df_daily['is_weekend'] = df_daily.index.dayofweek.isin([5, 6]).astype(int)  # Saturday/Sunday

    # Lag features (redemptions and sales)
    df_daily['lag_1'] = df_daily['redemptions'].shift(1)
    df_daily['lag_7'] = df_daily['redemptions'].shift(7)
    df_daily['lag_14'] = df_daily['redemptions'].shift(14)

    df_daily['sales_lag_1'] = df_daily['Sales Count'].shift(1)
    df_daily['sales_lag_7'] = df_daily['Sales Count'].shift(7)

    # Rolling averages
    df_daily['rolling_mean_3'] = df_daily['redemptions'].rolling(window=3).mean()
    df_daily['rolling_mean_7'] = df_daily['redemptions'].rolling(window=7).mean()

    # Percent change (trend indicator)
    df_daily['pct_change'] = df_daily['redemptions'].pct_change()

    # Cumulative total
    df_daily['cumulative_redemptions'] = df_daily['redemptions'].cumsum()

    # Fill missing values (after lags/rolling)
    df_daily.fillna(0, inplace=True)

    return df_daily

# Load data
file_path = r"C:\Users\jiwue\Downloads\Toronto Island Ferry Ticket Counts.csv"
df = load_data(file_path)

df.head(10)
```

## Technical Description Summary

his project focuses on predicting how many people will redeem ferry tickets or buy new ones in Toronto each day. The goal is to help the City of Toronto plan ferry services better by knowing how busy it will be in advance.

To tackle this, I built a custom Python tool called RedemptionModel. I designed it to be flexible and reusableâ€”something that could easily run different types of models, compare their results, and even generate visuals for better understanding. The dataset came from the City of Torontoâ€™s Open Data Portal. It contains daily records of ticket sales and redemptions. I cleaned and prepared the data, created new features like:

- Month

- Weekend flag

- Ticket usage from previous days (lags)

- Rolling averages

These features helped the models understand both short- and long-term patterns in ferry usage.

I experimented with five types of models:

1. Simple seasonal models that assume patterns repeat every year.

2. Season + trend models that add a general upward/downward direction to the seasonal cycle.

3. A sales-based version using decomposition on sales instead of redemptions.

4. XGBoost, a powerful machine learning model that picks up on deeper patterns.

5. LSTM, a deep learning model that learns from time sequences.

I used time-series cross-validation to train and test these models over multiple periods, ensuring results were realistic and reliable. I evaluated each model using three standard metrics: MAPE (percentage error), RMSE (average error), and RÂ² (how well the model explains the data).

Among all the methods, XGBoost clearly performed best. It consistently gave more accurate predictions, reducing RMSE by over 60% compared to the baseline, and achieved an RÂ² of 0.89â€”which means it could explain nearly 90% of what was going on in the data.
